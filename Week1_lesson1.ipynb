# 【問題1】100日目の米粒の数

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
todays_rice = 1
rice_count = 0
days = []
rice = []
rice_sum = []
for i in range(1, 101):
    day = i
    days.append(day)
    rice.append(todays_rice)
    rice_count += todays_rice
    rice_sum.append(rice_count)
    todays_rice = 2 ** day

plt.title("rice")
plt.xlabel("days")
plt.ylabel('rice')
plt.plot(days, rice) 
plt.show();

plt.title("rice_sum")
plt.xlabel("days")
plt.ylabel('rice_sum')
plt.plot(days, rice_sum) 
plt.show();
```

# 【問題2】100日目以外の米粒の数

```python
def compute_sorori_shinzaemon(day=100):
    """
    曽呂利新左衛門の米の逸話を検証する関数。日にちごとの米の粒の数を計算する。

    Parameteres
    --------------
    day: int
        米を貰う日数 (default : 100)

    Returns
    --------------
    list_n_grains : list
        ある日に貰う米の数のリスト（リストの長さはdayに等しくなる）
    list_total_grains : list
        ある日までに累計で貰う米の数のリスト（リストの長さはdayに等しくなる）
    """

    # ここにコードを書く。passは何もしないことを意味する命令である。
    list_n_grains = []
    list_total_grains = []
    todays_rice = 1
    rice_count = 0
    day_x = []
    
    for i in range(day + 1):
        days = i
        day_x.append(days)
        list_n_grains.append(todays_rice)
        rice_count += todays_rice
        list_total_grains.append(rice_count)
        todays_rice = 2 ** days

    return list_n_grains, list_total_grains , day_x


list_n_grains, list_total_grains, day_x = compute_sorori_shinzaemon(day=10)

plt.title("today_rice")
plt.xlabel("days")
plt.ylabel('list_n_grains')
plt.plot(day_x, list_n_grains) 
plt.show();

plt.title("rice_total")
plt.xlabel("days")
plt.ylabel('ist_total_grains')
plt.plot(day_x, list_total_grains) 
plt.show();
```

# 【問題3】何人が何日生活できる米か

```python
list_n_grains, list_total_grains, day_x = compute_sorori_shinzaemon(day=30)
total_grains = max(list_total_grains)

def survive(total_grains, man):
    """
    講義の日数でもらえる米粒の数を任意の人数で均等に消費した場合に何日生き残れるかの計算
    
    Parameteres
    --------------
    total_grains: int
        米粒の総数
    man: int
        大人の人数
        
    ness_rice: int
        大人一人当たりの１日必要な米数
    """
    ness_rice = 10000
    total_ness_rice = man * ness_rice
    suvive_days = total_grains // total_ness_rice
    print( man, '人では、',suvive_days, '日生活できます。', sep='')

print('米の総数：',total_grains, sep='')
survive(total_grains, 12)
    
```
