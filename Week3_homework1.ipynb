# 【問題4】irisデータセット自体の調査

## ・iris_setosa: ひおうぎあやめ
 高層湿原や湿った草地に群生し、高さは６０～７０センチになります。葉は剣状で、このかたちが檜扇に似ていることが名前の由来です。７月から８月ごろ、花茎の上部に２～３個の紫色の花を咲かせます。大きな外花被片は、「アヤメ」のような黄色地に紫色の網目模様があります。内花被片は小さく、卵状皮針形で先端が尖っているのが特徴です。

## ・iris_virginica: アイリスバージニカ
アイリス・バージニカは多年生植物です。植物には、2〜4個の直立またはアーチ型の明るい緑色の槍型の葉があります。がく片には、黄色から黄色がかったオレンジのスプラッシュがあります。1つの茎に2〜6個の花が咲きます。

## ・iris_virgicolr: アイリスベルシカラー
アイリスベルシカラーは、草本性の 多年生植物で、高さは10〜80 cm（4〜31インチ）成長します。それは、厚い根茎から大きな塊を形成する傾向があります。葉は中心で折り畳まれ、重なり合う平らな扇を形成します青い花には6 枚の花弁とがく片があります。

## 箱ひげ図とバイオリン図の違いとそれぞれの利点

### 2つの違い：　


箱ひげ図には平均/中央値と四分位範囲という要約統計量のみが含まれるのに対し、バイオリンプロットにはすべての分布データが含まれる。
