# 【問題5】Pythonの関数化

問題1から問題3までの範囲をまとめて行うPythonの関数を雛形を参考に作成してください。

そして、以下の3式の勾配を0.1間隔で求めてndarrayで出力してください。

```python
def compute_gradient(function, x_range=(-50, 50.1, 0.1)):
    """
    変化量を使い勾配を計算する。

    Parameters
    ----------------
    function : function
    求めたい関数、xのndarrayを入れるとyのndarrayを返すもの。
    x_range : tuple
    np.arange()の引数と同じ方法で範囲を指定する。

     Returns
     ----------------
    array_xy : ndarray, shape(n, 2)
    xとyを結合したもの。nはx_rangeに依存する。
    gradient : ndarray, shape(n-1,)
    関数の勾配。差をとるため要素は1つ減る
    """
    n, m, l = x_range
    array_x = np.arange(n, m , l)
    array_y = function(array_x)
    
    array_xy = np.vstack([array_x, array_y])
    array_xy = array_xy.T
    
    x_y_diff = array_xy[1:] - array_xy[:-1]
    gradient = x_y_diff[:,1] / x_y_diff[:, 0]
    
    return array_xy, gradient

def function1(array_x):

    array_x = array_x
    
    array_y = array_x  ** 2
    
    return array_y

array_xy, gradient = compute_gradient(function1)
```

```python
plt.title('function1')
plt.xlabel('x')
plt.ylabel('y')
plt.plot(array_xy[:, 0], array_xy[:, 1])
plt.show();
```

```python
plt.title('gradient_function1')
plt.xlabel('x')
plt.ylabel('gradient')
plt.plot(array_xy[1:, 0], gradient)
plt.show();
```

```python
def function2(array_x):

    array_x = array_x
    
    array_y = 2 * array_x**2 + 2**array_x
    
    return array_y

array_xy, gradient = compute_gradient(function2)
```

```python
plt.title('function2')
plt.xlabel('x')
plt.ylabel('y')
plt.plot(array_xy[:, 0], array_xy[:, 1])
plt.show();
```

```python
plt.title('gradient_function2')
plt.xlabel('x')
plt.ylabel('gradient')
plt.plot(array_xy[1:, 0], gradient)
plt.show();
```

```python
def function3(array_x):

    array_x = array_x
    
    array_y = np.sin(array_x**1/2)
    
    return array_y

array_xy, gradient = compute_gradient(function3, (0, 50.1,0.1))
```

```python
plt.title('function3')
plt.xlabel('x')
plt.ylabel('y')
plt.plot(array_xy[:, 0], array_xy[:, 1])
plt.show();
```

```python
plt.title('gradient_function3')
plt.xlabel('x')
plt.ylabel('gradient')
plt.plot(array_xy[1:, 0], gradient)
plt.show();
```

# 【問題6】最小値を求める
問題5で扱った関数のyの最小値を求めます。xの範囲も問題5と同様とします。

今回は勾配降下法ではなく、ndarrayのminメソッド、argminメソッドを用いてください。minメソッドは最小値を、argminメソッドは最小値のインデックスを取得することができます。

また、最小のyをとる前後の勾配を表示してください。

```python
def function1(array_x):

    array_x = array_x
    
    array_y = array_x  ** 2
    
    return array_y

array_xy, gradient = compute_gradient(function1)
```

```python
array_y = array_xy[:, 1]
y_min = array_y.min()
y_min_idx = array_y.argmin()
print("yの最小値:",y_min)
print("最小のyをとる前の勾配:",gradient[y_min_idx + 1])
print("最小のyをとる後の勾配:",gradient[y_min_idx - 1])
```

```python
array_y = array_xy[:, 1]
y_min = array_y.min()
y_min_idx = array_y.argmin()
print("yの最小値:",y_min)
print("最小のyをとる前の勾配:",gradient[y_min_idx + 1])
print("最小のyをとる後の勾配:",gradient[y_min_idx - 1])
```
