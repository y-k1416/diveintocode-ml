```python
import numpy as np
import matplotlib.pyplot as plt
```

# 【問題1】乱数の作成
平均が(-3, 0)、共分散行列が以下で表される2次元正規分布による乱数を500個作成してください。

```python
# 平均値
mu = [-3, 0]

# 共分散行列
sigm = [[1.0, 0.8], [0.8, 1.0]]

#　seed値設定し乱数を固定
np.random.seed(0)

# 2次元正規分布の乱数生成
random_scale = np.random.multivariate_normal(mu, sigm, 500)
random_scale
```

# 【問題2】散布図による可視化
問題1で作成したデータ点を散布図により可視化してください。

散布図はmatplotlibのplt.scatter()を使うことで描けます。

```python
x1 = random_scale[:,0]
x2 = random_scale[:,1]
plt.title('scatter')
plt.xlabel('x1')
plt.ylabel('x2')
plt.scatter(x1, x2)
plt.show()
```

# 【問題3】ヒストグラムによる可視化

```python
x1 = random_scale[:,0]
x2 = random_scale[:,1]

plt.title('hist_of_x1')
plt.xlabel('x1')
plt.xlim(-6, 3)
plt.ylabel('frequency')
plt.hist(x1)
plt.show()


plt.title('hist_of_x2')
plt.xlabel('x2')
plt.xlim(-6, 3)
plt.ylabel('frequency')
plt.hist(x2)
plt.show();
```

# 【問題4】データの追加
新たに平均が(0, -3)、共分散行列が以下で表される2次元正規分布による乱数を500個作成してください。
そして、問題1、4それぞれのデータをひとつの散布図として可視化してください。凡例として問題1のものは0、問題2のものは1を表示してください。

```python
# 平均値
mu2= [0, -3]

# 共分散行列
sigm2 = [[1.0, 0.8], [0.8, 1.0]]

#　seed値設定し乱数を固定
np.random.seed(0)

# 2次元正規分布の乱数生成
random_scale2 = np.random.multivariate_normal(mu2, sigm2, 500)
```

```python
x1 = random_scale[:,0]
x2 = random_scale[:,1]
y1 = random_scale2[:,0]
y2 = random_scale2[:, 1]
plt.title('scatter')
plt.xlabel('x1')
plt.ylabel('x2')
plt.scatter(x1, x2, label='0')
plt.scatter(y1, y2, label='1')
plt.legend()
plt.show()
```

# 【問題5】データの結合
データはまとめておいた方が後々扱いやすいです。問題1、4で作成したndarrayを 結合 し、(1000, 2)のndarrayとしてください。結合はnp.concatenate()やnp.vstack()を使うことで行えます。

```python
scale = np.vstack([random_scale, random_scale2])
scale.shape
```

```python
scale = np.concatenate([random_scale, random_scale2])
scale.shape
```

# 【問題6】ラベル付け
ひとまとめになった1000個のデータそれぞれに対して、問題1、4どちらで作成したものなのかを示す ラベル 付けを行ってください。問題1のものには0、問題4のものには1を対応させます。

0と1を含むラベルの列を新たに追加し、(1000, 3)のndarrayを作成してください。

機械学習に使用するデータセットはこのような形になっていることが多いです。

```python
scale1 = np.insert(random_scale, 2, 0, axis=1)
scale1
```

```python
scale2 = np.insert(random_scale2, 2, 1, axis=1)
scale2
```

```python
con_scale = np.vstack([scale1, scale2])
con_scale.shape
```
